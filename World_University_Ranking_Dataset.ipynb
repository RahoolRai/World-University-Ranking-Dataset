# üîπ Step-1 : Import all libraries which are required for dataset.

```python
# Import the all requires libraries for cleaning or preprocessing and Visualizations of World University Ranking Dataset.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# üîπStep-2 : Load & Explore Dataset of World University Ranking

```python
# Load the dataset of World University Ranking Dataset
df = pd.read_csv("World_University_Ranking_Dataset.csv")
df
```

# üîπStep-3 :üßπ Data Cleaning or Preprocessing of the World University Dataset
 The dataset was carefully cleaned by identifying and handling missing values in key numerical columns such as 'Citations', 'Teaching_Score', and 'International_Students_Percent' using mean imputation to ensure consistency and reliability for analysis.

```python
# Show the first five rows of World University Ranking Dataset
df.head()
```

```python
# Show the information of Column's data types and non-null counts of World University Ranking Dataset
df.info()
```

```python
# Number of unique values in each column of World University Ranking Dataset
df.nunique()
```

```python
# Show the statistics information of numeric columns of World University Ranking Dataset 
df.describe()
```

```python
# Check the missing values in each column of World University Ranking Dataset
print(df.isnull().sum())
```

```python
# Fill missing numerical values with the column mean 
df['Citations'].fillna(df['Citations'].mean(), inplace=True)
df['International_Students_Percent'].fillna(df['International_Students_Percent'].mean(), inplace=True)
df['Teaching_Score'].fillna(df['Teaching_Score'].mean(), inplace=True)
df
```

# üîπStep-4 : Verify all missing values are Handled and Should return 0 for all columns of World University Ranking Dataset 

```python
# Verify all missing values are Handled and Should return 0 for all columns of World University Ranking Dataset 
print(df.isnull().sum())  
```

# üìä Create the Visualizations include 7 different types of visualizations of World University Ranking Dataset which are given. 

# üìä Visualization Summary Of World University Ranking Dataset
---> A variety of visualizations including bar charts, line plots, histograms, heatmaps, pie charts, box plots, and scatter plots were used to explore and present insights from the dataset, uncovering trends in research output, international diversity, and university performance across countries.

# üìä 1. Bar Graph :  Top 10 Universities by Research Score
---> Shows the top universities globally based on research performance.
Helps compare which countries are leading in academic research output.

```python
# 1. Bar Graph: Top 10 Universities by Research Score
top10_research = df.nlargest(10, 'Research_Score')
sns.barplot(x='Research_Score', y='University_Name', data=top10_research, palette='viridis')
plt.title('Top 10 Universities by Research Score')
plt.xlabel('Research Score')
plt.ylabel('University Name')
plt.tight_layout()
plt.show()
```

# üìà 2. Line Graph : Average Research Score Over Time
---> Displays trends in research score across years.
Reveals improvements or declines in research productivity.

```python
# 2. Line Graph: World Rank vs Research Score (Top 50)
sns.lineplot(x='World_Rank', y='Research_Score', data=df.head(50), marker='o')
plt.title('World Rank vs Research Score (Top 50)')
plt.xlabel('World Rank')
plt.ylabel('Research Score')
plt.tight_layout()
plt.show()
```

# üìâ 3. Histogram Graph : Distribution of Teaching Scores
---> Shows the frequency of teaching scores across universities.
Helps identify how many institutions offer high-quality teaching.

```python
# 3. Histogram: Distribution of Student-Staff Ratio
sns.histplot(df['Student_Staff_Ratio'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Student-Staff Ratio')
plt.xlabel('Student-Staff Ratio')
plt.tight_layout()
plt.show()
```

# üî• 4. Heatmap Graph : Correlation Between Ranking Factors
---> Visualizes relationships between metrics like research, teaching, and citations.
Identifies which factors most influence overall university rankings.

```python
# 4. Heatmap: Correlation Between Numeric Features
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of University Metrics')
plt.tight_layout()
plt.show()
```

# ü•ß 5. Pie Chart : Distribution of Top Universities by Country
---> Displays how many top universities each country contributes.
Highlights dominant nations in higher education.

```python
# 5. Pie Chart: Distribution of Universities by Country (Top 5 Countries)
top_countries = df['Country'].value_counts().nlargest(5)
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 5 Countries by University Count')
plt.tight_layout()
plt.show()
```

# üì¶ 6. Box Plot : Spread of Citation Scores
---> Illustrates the distribution and outliers in citation scores.
Shows which institutions have extremely high academic influence.

```python
# 6. Box Plot: Research Score by Country (Top 5 Countries)
top5 = df[df['Country'].isin(df['Country'].value_counts().nlargest(5).index)]
sns.boxplot(x='Country', y='Research_Score', data=top5, palette='Set3')
plt.title('Research Score Distribution by Country')
plt.tight_layout()
plt.show()
```

# ‚ö´ 7. Scatter Plot : Teaching vs Research Score
---> Shows the relationship between teaching and research quality.
Reveals whether better teaching aligns with better research.

```python
# 7. Scatter Plot: Citations vs Research Score
sns.scatterplot(x='Citations', y='Research_Score', data=df, hue='Country', alpha=0.7)
plt.title('Citations vs Research Score')
plt.xlabel('Citations')
plt.ylabel('Research Score')
plt.tight_layout()
plt.show()
```
